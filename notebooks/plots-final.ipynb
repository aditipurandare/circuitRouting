```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Results for 100480 parameters
my_d = {'Runtime [s]': [11.0,
                        10.6, # From DreamPlace Paper (first entry of table I)
                        5.32, # Collected from actual run SGD TFCPU 10000 cells
                        5.5, # Collected from actual run ADAM TFCPU 10000 cells
                        1.48, # Collected from actual run SGD TFGPU 10000 cells
                        1.49], # Collected from actual run ADAM TFGPU 10000 cells
        'Framework Type ': ['PCB Enginner',
                    'RePlAce\nCPU\nIC Dataset',
                   'Tensorflow\nCPU',
                   'Tensorflow\nCPU',
                   'Tensorflow\nGPU-1080TI', 
                   'Tensorflow\nGPU-1080TI'],
        'Framework Type': ['PCB Enginner',
                           'RePlAce [1]\nCPU-40-Threads\nIC Dataset',
                           'Tensorflow\nCPU-12-Threads',
                           'Tensorflow\nCPU-12-Threads',
                           'Tensorflow\nGPU-1080TI', 
                           'Tensorflow\nGPU-1080TI'],
        'Optimizer': ['Manual',
                      'SGD',
                      'SGD',
                      'ADAM',
                      'SGD',
                      'ADAM'],
        'Number of Components': [427040/2,
                                 427040/2,
                                 427040/2,
                                 427040/2, 
                                 427040/2, 
                                 427040/2],
        'Number of Cores': [20,
                        20,
                         6,
                         6, 
                         1, 
                         1]
       }
```

```python
df = pd.DataFrame.from_dict(my_d)
df['Normalized Runtime [s]'] = df['Runtime [s]']*df['Number of Cores']
display(df)
```

```python
plt.figure(figsize=(16,8), dpi=80)
plt.rcParams.update({'font.size': 32})
ax = sns.barplot(x='Framework Type', y='Runtime [s]', hue='Optimizer', data=df)
#ax.set_title("(GP) 100 iterations of Cost+Update\n~213K Componenets")
ax.grid()
ax.tick_params(axis='both', which='major', labelsize=26)
#plt.tight_layout()
plt.show()
```
